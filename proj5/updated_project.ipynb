```python

#create a list of dictionaries

def make_dict(tables):
    databay_list = []

    for row in tables.findAll('tr'):
        d = {}
        tag = row.a
        title = tag
        if title:
            d['title'] = title.text
        else:
            title = ""
            d['title'] = title
        
        speaker = row.em
        if speaker:
            d['speaker'] = speaker.text
        else:
            speaker = ""
            d['speaker'] = speaker
        
        
        #link = row.a['href']
        
        try:
            d['link'] = "http://databythebay2016.sched.org" + row.a['href']
        except:
            link = ""
            d['link'] = link
        
        test = row.find('div', attrs = {"class" : "sched-description"})
        if test:
            d['describe'] = test.text
        else:
            test = ""
            d['describe'] = test

        databay_list.append(d) 

    return databay_list
```
